## Setting Threshold Value -- Based on the number of people that benefit from an existing water bottle station per station, we will set up a threshold value that would be applied to the buildings without water bottle station

```python
numberOfPeopleBenefitFromExistingWaterBottleStation = []
for key, value in existing.items():
  numberOfPeopleBenefitFromExistingWaterBottleStation.append(value)

numberOfPeopleThatWouldBenefitFromWaterBottleStation = []
for key, value in nonexisting.items():
  numberOfPeopleThatWouldBenefitFromWaterBottleStation.append(value)
```

```python
meanVal = sum(numberOfPeopleBenefitFromExistingWaterBottleStation) / len(numberOfPeopleBenefitFromExistingWaterBottleStation)
print(f"Average number of people from an existing water bottle station per station: {int(meanVal)}")

medianVal = statistics.median(numberOfPeopleBenefitFromExistingWaterBottleStation)
print(f"Median number of people from an existing water bottle station per station: {int(medianVal)}")

Q1 = statistics.quantiles(numberOfPeopleBenefitFromExistingWaterBottleStation, n=4, method='inclusive')[0]
Q3 = statistics.quantiles(numberOfPeopleBenefitFromExistingWaterBottleStation, n=4, method='inclusive')[2]
IQR = Q3 - Q1

print(f"Interquartile range from an existing water bottle station per station: {int(IQR)}")
```

```python
greaterThanMean = []
greaterThanMedian = []
for x in numberOfPeopleThatWouldBenefitFromWaterBottleStation:
  if x > meanVal:
    greaterThanMean.append(x)
  if x > medianVal:
    greaterThanMedian.append(x)

greaterThanMeanDictionary = {}
for x in greaterThanMean:
  for key, value in nonexisting.items():
    if value == x:
      greaterThanMeanDictionary[key] = value

greaterThanMedianDictionary = {}
for x in greaterThanMedian:
  for key, value in nonexisting.items():
    if value == x:
      greaterThanMedianDictionary[key] = value
```
